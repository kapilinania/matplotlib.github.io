```python
# Assuming 'data' is your DataFrame with 'race/ethnicity', 'parental level of education', 'math score', 'reading score', and 'writing score' columns
gender_education_scores = data.groupby(['race/ethnicity', 'parental level of education'])[['math score', 'reading score', 'writing score']].count()

# Add a new column 'total' that represents the sum of the three scores
gender_education_scores['total'] = gender_education_scores['math score'] + gender_education_scores['reading score'] + gender_education_scores['writing score']

# Print the resulting DataFrame
gender_education_scores
```

```python
# Reset the index to create a single string label for the x-axis
gender_education_scores.reset_index(inplace=True)

# Create the bar plot
plt.figure(figsize=(12, 6))  # Set the figure size
plt.bar(range(len(gender_education_scores)), gender_education_scores['total'])

# Set the x-axis labels and titles
plt.xlabel('Race/Ethnicity and Parental Education Level')
plt.ylabel('Total Score')
plt.title('Total Score by Race/Ethnicity and Parental Education Level')

# Customize the x-axis labels with the combined 'race/ethnicity' and 'parental level of education'
x_labels = [f"{race}, {education}" for race, education in zip(gender_education_scores['race/ethnicity'], gender_education_scores['parental level of education'])]
plt.xticks(range(len(gender_education_scores)), x_labels, rotation=45, ha='right')

# Show the plot
plt.show()
```
