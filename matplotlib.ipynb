#Matplotlib
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.


##Installation of Matplotlib

```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([5,10])
y=np.array([15,10])
plt.plot(x, y, label='Line 1', marker='o')
# Add labels to the x and y axes
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Add a legend
plt.legend()

# Add a title
plt.title('Line Plot Example')
plt.show()
```

###Plotting x and y points
Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
# The x-axis is the horizontal axis.

# The y-axis is the vertical axis.
x= np.array([1,5])
y= np.array([3,10])
plt.plot(x,y, label='Line 1', marker='*')
# Add labels to the x and y axes
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Add a legend
plt.legend()

# Add a title
plt.title('Line Plot Example')
plt.show()
```

```python
# Data for the plots
x = np.linspace(0, 2 * np.pi, 100)  # Create an array of values from 0 to 2*pi
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
y5 = np.log(x)
y6 = np.sqrt(x)

# Create a 2x3 grid of subplots
plt.figure(figsize=(12, 8))  # Adjust the figure size
plt.subplots_adjust(hspace=0.5)  # Adjust vertical spacing

# Plot 1: Sine
plt.subplot(2, 3, 1)
plt.plot(x, y1)
plt.title('Sine')

# Plot 2: Cosine
plt.subplot(2, 3, 2)
plt.plot(x, y2)
plt.title('Cosine')

# Plot 3: Tangent
plt.subplot(2, 3, 3)
plt.plot(x, y3)
plt.title('Tangent')

# Plot 4: Exponential
plt.subplot(2, 3, 4)
plt.plot(x, y4)
plt.title('Exponential')

# Plot 5: Logarithm
plt.subplot(2, 3, 5)
plt.plot(x, y5)
plt.title('Natural Logarithm')

# Plot 6: Square Root
plt.subplot(2, 3, 6)
plt.plot(x, y6)
plt.title('Square Root')

#supertitle
plt.suptitle("kapil wali maths")
# Show the subplots
plt.show()
```

```python
# Angles in degrees from 0 to 360
angles = np.linspace(0, 360, 1000)
radians = np.deg2rad(angles)  # Convert angles to radians

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
fig.suptitle('Trigonometric Functions', fontsize=16)

# Function names for titles
function_names = ['Sine', 'Cosine', 'Tangent', 'Cotangent', 'Secant', 'Cosecant']

for i, ax in enumerate(axes.flat):
    function_name = function_names[i]

    if function_name == 'Tangent':
        y = np.tan(radians)
    elif function_name == 'Cotangent':
        y = 1 / np.tan(radians)
    elif function_name == 'Secant':
        y = 1 / np.cos(radians)
    elif function_name == 'Cosecant':
        y = 1 / np.sin(radians)
    else:
        y = np.sin(radians) if function_name == 'Sine' else np.cos(radians)

    ax.plot(angles, y)
    ax.set_title(f'{function_name}')

# Adjust the spacing between subplots
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the subplots
plt.show()
```

##Size
You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

##Alpha

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

#Matplotlib Pie Charts
With Pyplot, you can use the pie() function to draw pie charts:

```python
y = np.array([4,10,6,15,20])
plt.pie(y)
plt.show()
```

##Labels

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

##Start Angle
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 160)
plt.show()
```

##Explode

Each value represents how far from the center each wedge is displayed:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.8, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

##Shadow
Add a shadow to the pie chart by setting the shadows parameter to True:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

##Colors
You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

##Legend
To add a list of explanation for each wedge, use the legend() function:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
#Legend With Header
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.legend(title = "Four Fruits:")
plt.show()
```
